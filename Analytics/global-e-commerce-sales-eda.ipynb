# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Revenue', data=df)
plt.title(f'{plot_no}. Regional Revenue Comparison Shows Market Strength')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Customer_Segment', data=df)
plt.title(f'{plot_no}. Customer Segment Distribution Reveals Target Audience')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Revenue', y='Profit', hue='Category', data=df)
plt.title(f'{plot_no}. Revenue vs Profit Relationship Across Categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(x='Month', y='Revenue', data=df, estimator='mean')
plt.title(f'{plot_no}. Monthly Revenue Trend Shows Seasonality Patterns')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Shipping_Method', y='Shipping_Cost', data=df)
plt.title(f'{plot_no}. Shipping Cost Comparison by Method')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Discount', y='Revenue', data=df)
plt.title(f'{plot_no}. Impact of Discount on Revenue Generation')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Order_Status', data=df)
plt.title(f'{plot_no}. Order Status Distribution Indicates Fulfillment Health')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Season', y='Profit', data=df)
plt.title(f'{plot_no}. Seasonal Profit Trends Reveal Peak Business Periods')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Shipping_Days', y='Profit', data=df)
plt.title(f'{plot_no}. Shipping Time vs Profit Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(df[['Revenue','Cost','Profit','Discount','Shipping_Cost','Shipping_Days']].corr(), annot=True)
plt.title(f'{plot_no}. Correlation Heatmap of Key Business Metrics')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Category', y='Quantity', data=df)
plt.title(f'{plot_no}. Product Category vs Quantity Sold Analysis')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Cost', y='Profit', data=df)
plt.title(f'{plot_no}. Cost vs Profit Efficiency Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='Region', y='Profit_Margin_%', data=df)
plt.title(f'{plot_no}. Profit Margin Distribution Across Regions')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Quarter', y='Revenue', data=df)
plt.title(f'{plot_no}. Quarterly Revenue Performance Comparison')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Unit_Price', y='Quantity', data=df)
plt.title(f'{plot_no}. Price vs Quantity Sold Demand Analysis')
show_fig()
plot_no += 1
```
